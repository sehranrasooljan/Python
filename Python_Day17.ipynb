```python
# WAP to get infromation of a student in terms of 
# semseter number and marks obtained
# create a dataframe of collected info and plot it in a graph
# x axis as semester number and y axis as obtained marks

SemCnt=int(input("Enter Total Sems: "))
Sem=[]
Mrks=[]
for i in range(SemCnt):
  Sem.append(i+1)
  Mrks.append(int(input("Enter Obtained Marks in Sem "+str(i+1)+": ")))
Data=pd.DataFrame({"Semester":Sem,"Marks":Mrks})
print(Data)
```

```python
plt.plot(Data["Semester"],Data["Marks"],"r:d")
plt.show()
```

```python
Data=pd.DataFrame(np.random.randint(10,100,(6,5)),
                  columns=["St1","St2","St3","St4","St5"])
Data.index=["Sem1","Sem2","Sem3","Sem4","Sem5","Sem6"]
print(Data)
```

```python
plt.plot([1,2,3,4,5,6],Data["St1"],"-ro")
plt.plot([1,2,3,4,5,6],Data["St2"],"-bo")
plt.plot([1,2,3,4,5,6],Data["St3"],"-rd")
plt.plot([1,2,3,4,5,6],Data["St4"],"-rs")
plt.plot([1,2,3,4,5,6],Data["St5"],"-bs")
plt.title("Graphs of Performance")
plt.xlabel("semesters")
plt.ylabel("Obt Marks")
plt.legend(["St1","St2","St3","St4","St5"])
plt.show()
```
